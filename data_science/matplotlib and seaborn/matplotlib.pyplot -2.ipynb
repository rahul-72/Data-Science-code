```python
deg = np.arange(0,361,15)
deg
```

```python
rad = np.deg2rad(deg)
rad
```

```python
sin = np.sin(rad)
sin
```

```python
cos = np.cos(rad)
cos
```

```python
plt.plot(deg,sin)
plt.plot(deg,cos)
```

```python
plt.scatter(rad,cos)
plt.scatter(rad,sin)
```

```python
plt.plot(rad)
plt.plot(rad**2)
plt.plot(rad**3)
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2  = fig.add_axes([0.11,0.5,0.633,0.4])

ax1.plot(rad,rad,'r',label='1D',lw='3')
ax1.plot(rad,rad**2,'g',label='2D',lw='3')
ax1.plot(rad,rad**3,'b',label='3D',lw='3',alpha=0.9)
ax1.set_title("x, x_seq, x_cube")
ax1.legend(loc=7)
ax1.set_xlabel('Radians')
ax1.set_ylabel('Values')

ax2.plot(deg,sin,c='r',lw=2,label='sin')
ax2.plot(deg,cos,c='g',lw=2,label='cos')
ax2.plot(deg,[0]*25,'k:',lw=2)
ax2.set_xticks(np.arange(0,361,30))
ax2.set_yticks(np.arange(-1,1.1,0.2))
ax2.legend()
ax2.set_title("sin & cos")
ax2.set_xlabel("Angles")
ax2.set_ylabel("Values")
ax2.grid()
ax1.grid()
plt.show()
```

```python
# Graph or Plots 
x = np.array([ 3,4,5,3,5,4,3,2,1,8,])
y = np.array([ 45,50,65,40,70,65,50,20,10,80])
plt.scatter(x,y)
plt.plot([x.mean()]*len(x),y,'r')
plt.plot(x,[y.mean()]*len(y),'g')
for var in range(len(x)) : 
    plt.plot([x[var],x.mean()],[y[var],y[var]],'r:',alpha=1)
    plt.plot([x[var],x[var]],[y[var],y.mean()],'g:')
plt.show()
```
