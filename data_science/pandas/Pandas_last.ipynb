```python
cricket = pd.DataFrame({
    "Name" : [ 'ram','shyam','ghyanshyam','hari','mohan','sachin'],
    "Category" : pd.Categorical(['bad','excellent','good','avg','avg','excellent']),
    
})
```

```python
cricket
```

```python
football
```

```python
pd.merge(cricket,football,on='Name') # inner join 
```

```python
pd.merge(cricket,football,on='Name',how='left') # left join 
```

```python
pd.merge(cricket,football,on='Name',how='right') # right join 
```

```python
pd.merge(cricket,football,on='Name',how='outer') # full join 
```

```python
pd.merge(cricket,football,left_on="Name",right_on="stname",how='outer') # full join 
```
