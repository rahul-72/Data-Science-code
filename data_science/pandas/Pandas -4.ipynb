#### Q.1 which team has won maximum number ipl matches ? 

```python
df.head()
```

```python
winner = df['winner']
```

```python
type(winner)
```

```python
winner.head()
```

```python
# value_counts --> Categorical 
most_wins = winner.value_counts()
```

```python
most_wins
```

```python
most_wins.describe()
```

```python
plt.style.use('ggplot')
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1,1,1)
ax.set_title("Most Winning Team")
ax.set_yticks(np.arange(0,110,5))
ax.set_xlabel('Team Names')
ax.set_ylabel("Frequency")
most_wins.plot(kind='bar',ax=ax,label='winner')
plt.legend()
plt.show()

sns.countplot(df['winner'])
plt.xticks(rotation='vertical')
plt.show()
```

Q. which has won most mathes batting first alse had win the toss ? 

```python
df.columns
```

```python
df['toss_decision'].head()
```

```python
new_df = df[df['toss_decision']=='bat'][['toss_winner','winner']]
```

```python
new_df.reset_index(drop=True,inplace=True)
new_df.head()
```

```python
winner_bat = new_df[ new_df['toss_winner'] == new_df['winner'] ]
```

```python
plt.style.use('ggplot')
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1,1,1)
ax.set_title("Most Math hosting Venu")
ax.set_yticks(np.arange(0,110,5))
ax.set_xlabel('Venu Name')
ax.set_ylabel("Frequency")
data = winner_bat['winner'].value_counts()
data.plot(kind='bar',ax=ax,label='venues')
plt.legend()
plt.show()
```
