```python
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns 
```

```python
plt.style.use('seaborn')
```

```python
%matplotlib inline 
```

```python
plt.scatter(np.arange(10),np.random.randn(10),s=100)
plt.show()
```

```python
plt.plot(np.arange(10),np.random.randn(10))
plt.show()
```

```python
plt.hist(np.random.normal(45,10,1000),bins=50)
plt.show()
```

```python
years = [ 2015,2016,2017,2018,2019]
growth = [ 5.6,7.8,9.0,6.7,7]
color = [ 'blue','green','red','cyan','magenta']
plt.bar(years,height=growth,color=color)
plt.savefig('growth.jpg')
plt.show()
```

```python
years = [ 2015,2016,2017,2018,2019]
growth = [ 5.6,7.8,9.0,6.7,7]
color = [ 'blue','green','red','cyan','magenta']
plt.pie(x=growth,labels=years,colors=color,autopct='%.2f')
plt.show()
```

```python
plt.boxplot(np.random.randint(1,10,100))
plt.show()
```

```python
# distributions 
```

```python
data = np.random.normal(45,5,1000)
fig = plt.figure(figsize=(10,5))
ax = fig.add_axes([1,1,1,1])
sns.distplot(data,ax=ax,kde=False)
plt.show()
```

```python
tips = sns.load_dataset('tips') # pandas data frame
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'],kde=False)
plt.show()
```

```python
sns.distplot(tips['tip'],kde=False)
plt.show()
tips.columns
```

```python
sns.pairplot(tips,hue='smoker')
plt.show()
```

```python
tips.shape
```

```python
=
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,color='red',kind='hex') # scatter
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,color='red',kind='kde') # scatter
plt.show()
```

```python
sns.rugplot(tips['tip'])
plt.show()
```

```python
sns.violinplot(y='sex',x='total_bill',hue='day',data=tips,orient='h')
plt.show()
```

```python
sns.stripplot(x='sex',y='total_bill',data=tips,hue='time',palette='rainbow')
plt.show()
```

```python
sns.swarmplot(x='sex',y='total_bill',data=tips,hue='time',palette='rainbow')
sns.violinplot(x='sex',y='total_bill',data=tips,hue='time',palette='rainbow',split=True)
plt.show()
```

```python
sns.boxplot(x='sex',y='total_bill',data=tips,hue='smoker',palette='rainbow')
plt.show()
```

```python
sns.boxenplot(x='sex',y='total_bill',data=tips,hue='smoker',palette='rainbow')
plt.show()
```

```python
sns.countplot(x='sex',data=tips)
plt.show()
```

```python
sns.countplot(x='smoker',data=tips,hue='sex',palette='coolwarm')
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
plt.show()
```

```python
tips['total_bill'].mean()
```

```python
tips.query('sex=="Male"')['total_bill'].mean()
```

```python
tips.groupby('sex').mean()
```

```python
fig = plt.figure(figsize=(5,2))
for var in [ 'box','strip','swarm','violin'] : 
    sns.factorplot(x='sex',y='total_bill',data=tips,kind=var)
plt.show()
```

```python
flights = sns.load_dataset('flights')
```

```python
print(flights.shape)
print(flights['year'].unique())
flights.head()
```

```python
nt = flights.pivot(values="passengers",index="month",columns="year")
```

```python
nt.head()
```

```python
sns.heatmap(nt,cmap='coolwarm',linecolor='white',lw=2)
plt.show()
```

```python
sns.clustermap(nt,cmap='coolwarm',linecolor='white',lw=2)
plt.show()
```
