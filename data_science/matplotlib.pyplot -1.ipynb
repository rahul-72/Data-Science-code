```python
# Graph or Plots 
height = [ 3,4,5,3,5,4,3,2,1,8]
weight = [ 45,50,65,40,70,65,50,20,10,80]
x = height
y = weight
```

```python
len(x) == len(y)
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
plt.figure(figsize=(10,5))
plt.plot(x,y,label='Line')
plt.scatter(x,y,s=100,c='r',alpha=0.7,marker='H',label='Scatter')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title("Relation Between Height and Weight")
plt.xticks([var for var in range(11) ])
plt.yticks([ var for var in range(0,101,5)])
plt.grid()
plt.legend()
plt.show()
```

```python
# sin(radians) 
# angle 0 15 30 45 60 
# radin

angles = np.arange(0,361,15)
```

```python
radian = np.deg2rad(angles)
radian
```

```python
sin = np.sin(radian)
sin
```

```python
# plot sin and cos scatter plots in same figure
# x-ticks  should be 0,15,30,45,.....
# y-ticks -1,-0.8,-0.6,....0,0.2,0.4.....
cos = np.cos(radian)
```

```python
plt.plot(angles,sin)
plt.scatter(angles,cos,c='r')
```

```python
plt.figure(figsize=(10,5))
deg = np.arange(0,361,15)
rad = np.deg2rad(deg)
sin = np.sin(rad)
cos = np.cos(rad)
plt.scatter(angles,sin,label='sin')
plt.plot(angles,sin,'b:',alpha=0.5)
plt.scatter(angles,cos,label='cos')
plt.plot(angles,cos,'r:')
plt.legend()
plt.xticks(angles)
plt.yticks(np.arange(-1,1.1,.2))
plt.title('Sin & Cos')
plt.xlabel('Angles')
plt.ylabel('Value')
plt.grid()
plt.show()
```
